```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_ftir_csv(file):
    df = pd.read_csv(file)

    # Nos quedamos solo con las dos primeras columnas
    df = df.iloc[:, :2]
    df.columns = ["wavenumber", "signal"]

    return df
```

```python
def plot_ftir_comparison(
    file_paa,
    file_nopaa,
    mode="abs",
    wn_min=800,
    wn_max=4000
):
    df_paa   = load_ftir_csv(file_paa)
    df_nopaa = load_ftir_csv(file_nopaa)

    df_paa   = df_paa[(df_paa["wavenumber"] >= wn_min) & (df_paa["wavenumber"] <= wn_max)]
    df_nopaa = df_nopaa[(df_nopaa["wavenumber"] >= wn_min) & (df_nopaa["wavenumber"] <= wn_max)]

    plt.figure(figsize=(8,5))

    # No-PAA → verde fuerte
    plt.plot(
        df_nopaa["wavenumber"],
        df_nopaa["signal"],
        color="#2ca02c",
        linewidth=1.8,
        label="No-PAA"
    )

    # PAA → naranja fuerte
    plt.plot(
        df_paa["wavenumber"],
        df_paa["signal"],
        color="#ff7f0e",
        linewidth=1.8,
        label="PAA"
    )

    plt.gca().invert_xaxis()

    plt.xlabel("Número de onda (cm⁻¹)")
    plt.ylabel("Absorbancia (a.u.)" if mode == "abs" else "Transmitancia (%)")

    plt.title(
        f"FTIR3 – Comparación PAA vs No-PAA ({'Absorbancia' if mode == 'abs' else 'Transmitancia'})"
    )

    plt.legend()
    plt.tight_layout()
    plt.show()
```
