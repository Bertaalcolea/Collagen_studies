```python
import pandas as pd
import matplotlib.pyplot as plt

def load_ftir_csv(file):
    df = pd.read_csv(file)
    df = df.iloc[:, :2]
    df.columns = ["wavenumber", "signal"]
    return df
```

```python
files_nopaa_abs = [
    "non-paa ovine lyo tube 1 of 4 (abs)12-08-2025_2025-12-08T10-52-30.csv",
    "non-paa ovine lyo tube 1 of 4 (abs) 12-08-2025 (2)_2025-12-08T10-54-04.csv",
    "non-paa ovine lyo tube 1 of 4 (abs) 12-08-2025 (3)_2025-12-08T11-01-32 (1).csv"
]

greens = ["#1b7837", "#5aae61", "#a6dba0"]

plt.figure(figsize=(9,5))

for i, f in enumerate(files_nopaa_abs):
    df = load_ftir_csv(f)
    df = df[(df["wavenumber"] >= 800) & (df["wavenumber"] <= 4000)]

    plt.plot(
        df["wavenumber"],
        df["signal"],
        color=greens[i],
        linewidth=1.8,
        label=f"No-PAA (rep {i+1})"
    )

# ðŸ”‘ Ejes FIJOS
plt.xlim(4000, 800)
plt.ylim(0, 0.30)


plt.xlabel("NÃºmero de onda (cmâ»Â¹)")
plt.ylabel("Absorbancia (a.u.)")
plt.title("FTIR â€“ No-PAA (Absorbancia) â€“ 3 rÃ©plicas")

plt.legend()
plt.tight_layout()
plt.show()
```

```python
files_paa_abs = [
    "paa ovine lyo tube 4 of 4 (abs) 12-08-2025 (1)_2025-12-08T11-16-16.csv",
    "paa ovine lyo tube 4 of 4 (abs) 12-08-2025 (3)_2025-12-08T11-26-05.csv",
    "paa ovine lyo tube 4 of 4 (abs) 12-08-2025 (4)_2025-12-08T16-50-36.csv"
]

oranges = ["#e6550d", "#fd8d3c", "#fdae6b"]

plt.figure(figsize=(9,5))

for i, f in enumerate(files_paa_abs):
    df = load_ftir_csv(f)
    df = df[(df["wavenumber"] >= 800) & (df["wavenumber"] <= 4000)]

    plt.plot(
        df["wavenumber"],
        df["signal"],
        color=oranges[i],
        linewidth=1.8,
        label=f"PAA (rep {i+1})"
    )

# ConvenciÃ³n FTIR
plt.gca().invert_xaxis()



plt.xlabel("NÃºmero de onda (cmâ»Â¹)")
plt.ylabel("Absorbancia (a.u.)")
plt.title("FTIR â€“ PAA (Absorbancia) â€“ 3 rÃ©plicas")

plt.legend()
plt.tight_layout()
plt.show()
```
